```python
# NOW READ IN 2023

zip_path_2023 = "../data/raw/voa_2023_list.zip"

list_entry_cols = [
    "billing_authority_code",  # 1:  0230
    "ndr_community_code",      # 2:  (empty)
    "ba_reference_number",     # 3:  201002510000101
    "primary_desc_code",       # 4:  NX
    "primary_desc_text",       # 5:  INDEPENDENT DISTRIBUTION...
    "uarn",                    # 6:  14675610000
    "full_property_identifier",# 7:  INDEPENDENT DISTRIBUTION...LUTON...
    "firm_name",               # 8:  ESP ELECTRICITY LTD
    "number_or_name",          # 9:  INDEPENDENT DISTRIBUTION...
    "street",                  # 10: ARMSTRONG ROAD
    "town",                    # 11: LUTON
    "postal_district",         # 12: (empty)
    "county",                  # 13: BEDS
    "postcode",                # 14: LU2 0FY
    "effective_date",          # 15: 01-APR-2026
    "composite_indicator",     # 16: (empty)
    "rateable_value",          # 17: 475
    "appeal_settlement_code",  # 18: (empty)
    "assessment_reference",    # 19: 29952695000
    "list_alteration_date",    # 20: (empty)
    "scat_code_and_suffix",    # 21: 742U
    "sub_street_3",            # 22: (empty)
    "sub_street_2",            # 23: (empty)
    "sub_street_1",            # 24: (empty)
    "case_number",             # 25: 39957812282
    "current_from_date",       # 26: 01-APR-2026
    "current_to_date",         # 27: (empty)
    "trailing",                # 28: artefact of line ending with *
]

with zipfile.ZipFile(zip_path_2023) as z:
    with z.open(z.namelist()[0]) as f:
        df_2023 = pd.read_csv(
            f,
            sep="\\*",
            header=None,
            dtype=str,
            encoding="latin-1",
            engine="python"
        )

# Drop row counter (col 0) and trailing artefact, assign names
df_2023 = df_2023.drop(columns=[0, 28])
df_2023.columns = list_entry_cols[:-1]  # drop "trailing" from names too

# Sanity check
print(df_2023[["billing_authority_code", "uarn", "rateable_value", "scat_code_and_suffix", "town"]].head())
```

```python
# Clean up both dataframes (2026 and 2023)

# Convert RV to numeric in both
df["rateable_value"] = pd.to_numeric(df["rateable_value"], errors="coerce")
df_2023["rateable_value"] = pd.to_numeric(df_2023["rateable_value"], errors="coerce")

# Extract numeric SCat code (strip suffix letter) in both
df["scat_code"] = df["scat_code_and_suffix"].str[:3].str.strip()
df_2023["scat_code"] = df_2023["scat_code_and_suffix"].str[:3].str.strip()

print("2026 RV sample:", df["rateable_value"].describe())
print("2023 RV sample:", df_2023["rateable_value"].describe())
```

```python
# Load and apply the RHL crosswalk to both

crosswalk = pd.read_excel(
    "../data/reference/SCat_RHL_Crosswalk_2026.xlsx",
    sheet_name="SCat RHL Crosswalk",
    dtype=str
)

# Remove section header rows
crosswalk = crosswalk[~crosswalk["SCat Code"].str.startswith("---", na=True)]
crosswalk = crosswalk[["SCat Code", "RHL Category", "RHL Status"]].rename(
    columns={"SCat Code": "scat_code"}
)

# Apply to both lists
df = df.merge(crosswalk, on="scat_code", how="left")
df_2023 = df_2023.merge(crosswalk, on="scat_code", how="left")

df["RHL Status"] = df["RHL Status"].fillna("UNCERTAIN")
df_2023["RHL Status"] = df_2023["RHL Status"].fillna("UNCERTAIN")
```

```python
# Define multiplier functions

def multiplier_2026(row):
    rv = row["rateable_value"]
    rhl = row["RHL Status"]
    if pd.isna(rv):
        return None
    if rv >= 500000:
        return 0.508
    if rhl == "YES":
        return 0.382 if rv < 51000 else 0.400
    return 0.432 if rv < 51000 else 0.480

def multiplier_2023(row):
    rv = row["rateable_value"]
    if pd.isna(rv):
        return None
    # Current system: two multipliers only, based on RV
    return 0.499 if rv < 51000 else 0.554

df["multiplier_2026"] = df.apply(multiplier_2026, axis=1)
df["estimated_bill_2026"] = df["rateable_value"] * df["multiplier_2026"]

df_2023["multiplier_2023"] = df_2023.apply(multiplier_2023, axis=1)
df_2023["estimated_bill_2023"] = df_2023["rateable_value"] * df_2023["multiplier_2023"]
```

```python
# Join the two lists on UARN

# Keep only the columns we need from each list
df_2026_slim = df[[
    "uarn", "rateable_value", "scat_code", "RHL Status", "RHL Category",
    "multiplier_2026", "estimated_bill_2026",
    "billing_authority_code", "town", "postcode", "primary_desc_text"
]].rename(columns={"rateable_value": "rv_2026"})

df_2023_slim = df_2023[[
    "uarn", "rateable_value", "multiplier_2023", "estimated_bill_2023"
]].rename(columns={"rateable_value": "rv_2023"})

# Join on UARN
df_combined = df_2026_slim.merge(df_2023_slim, on="uarn", how="left")

# Calculate changes
df_combined["rv_change"] = df_combined["rv_2026"] - df_combined["rv_2023"]
df_combined["rv_change_pct"] = (df_combined["rv_change"] / df_combined["rv_2023"]) * 100
df_combined["bill_change"] = df_combined["estimated_bill_2026"] - df_combined["estimated_bill_2023"]
df_combined["bill_change_pct"] = (df_combined["bill_change"] / df_combined["estimated_bill_2023"]) * 100

print(f"Matched properties: {df_combined['rv_2023'].notna().sum():,}")
print(f"New in 2026 (no 2023 match): {df_combined['rv_2023'].isna().sum():,}")
print(df_combined.head())
```

```python
# Safe percentage calculations
df_combined["bill_change_pct"] = (
    df_combined["bill_change"]
    .div(df_combined["estimated_bill_2023"])
    .replace([float("inf"), float("-inf")], None)
    .multiply(100)
)

df_combined["rv_change_pct"] = (
    df_combined["rv_change"]
    .div(df_combined["rv_2023"])
    .replace([float("inf"), float("-inf")], None)
    .multiply(100)
)

# Diagnostics
print(f"Zero/null 2023 bills: {(df_combined['estimated_bill_2023'].isna() | (df_combined['estimated_bill_2023'] == 0)).sum():,}")
print(f"Properties excluded from pct calculation: {df_combined['bill_change_pct'].isna().sum():,}")
print()

# Summary
print("=== BILL CHANGE SUMMARY BY RHL CATEGORY ===")
print(df_combined.groupby("town").agg(
    properties=("uarn", "count"),
    total_bill_2023=("estimated_bill_2023", "sum"),
    total_bill_2026=("estimated_bill_2026", "sum"),
    avg_bill_change_pct=("bill_change_pct", "mean")
).round(1))
```

```python
type_summary = df_combined.groupby(["RHL Category", "scat_code"]).agg(
    properties=("uarn", "count"),
    total_rv_2023=("rv_2023", "sum"),
    total_rv_2026=("rv_2026", "sum"),
    avg_rv_change_pct=("rv_change_pct", "mean"),
    avg_bill_change_pct=("bill_change_pct", "mean"),
    total_bill_change=("bill_change", "sum")
).sort_values("avg_bill_change_pct")

# Biggest winners (bill decrease)
#print("=== BIGGEST BILL DECREASES BY SCAT ===")
#print(type_summary.head(10))

# Biggest losers (bill increase)
print("=== BIGGEST BILL INCREASES BY SCAT ===")
print(type_summary.tail(10))
```
